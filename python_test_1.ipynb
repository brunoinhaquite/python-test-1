Questão 1 — Classificação de Temperaturas

```python
consumo = [120, 95, 200, 180, 75, 220, 160]
classificacoes_consumo = []
dias_alto_consumo = 0

for consumo_diario in consumo:
    if consumo_diario < 100:
        classificacoes_consumo.append("Baixo")
    elif 100 <= consumo_diario <= 180:
        classificacoes_consumo.append("Moderado")
    else: # consumo_diario > 180
        classificacoes_consumo.append("Alto")
        dias_alto_consumo += 1

total_consumo_semana = sum(consumo)
media_consumo_semana = total_consumo_semana / len(consumo)

print(f"Consumo diário original: {consumo}")
print(f"Classificações de consumo: {classificacoes_consumo}")
print(f"Total de consumo da semana: {total_consumo_semana} kWh")
print(f"Média semanal de consumo: {media_consumo_semana:.2f} kWh")

if dias_alto_consumo > 2:
    print(f"ALERTA: Houve {dias_alto_consumo} dias com consumo alto, o que excede o limite de 2 dias.")
```

```python
temperaturas = [18, 22, 30, 35, 28, 15, 40]
classificacoes = []

# Initialize counters for each category
contador_frio = 0
contador_agradavel = 0
contador_quente = 0

# Classify each temperature and update counters
for temp in temperaturas:
    if temp < 20:
        classificacoes.append("Frio")
        contador_frio += 1
    elif 20 <= temp <= 30:
        classificacoes.append("Agradável")
        contador_agradavel += 1
    else: # temp > 30
        classificacoes.append("Quente")
        contador_quente += 1

print(f"Temperaturas originais: {temperaturas}")
print(f"Classificações: {classificacoes}")
print(f"\nContagem por categoria:")
print(f"Frio: {contador_frio} dias")
print(f"Agradável: {contador_agradavel} dias")
print(f"Quente: {contador_quente} dias")
```

## Questão 2 — Sistema de Avaliação de Alunos

(0,5 ponto)

Dada a lista de notas finais:

`notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]`

Classifique:

*   `< 5` → "Reprovado"
*   `5 a 7` → "Recuperação"
*   `> 7` → "Aprovado"

**Tarefas:**

*   Gerar lista de classificações
*   Calcular:
    *   Quantidade de aprovados
    *   Percentual de reprovados

```python
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]
classificacoes_notas = []

# Initialize counters
aprovados = 0
reprovados = 0

# Classify each grade and update counters
for nota in notas:
    if nota < 5:
        classificacoes_notas.append("Reprovado")
        reprovados += 1
    elif 5 <= nota <= 7:
        classificacoes_notas.append("Recuperação")
    else: # nota > 7
        classificacoes_notas.append("Aprovado")
        aprovados += 1

# Calculate total students and percentage of failing students
total_alunos = len(notas)
percentual_reprovados = (reprovados / total_alunos) * 100

print(f"Notas originais: {notas}")
print(f"Classificações das notas: {classificacoes_notas}")
print(f"\nQuantidade de alunos aprovados: {aprovados}")
print(f"Percentual de reprovados: {percentual_reprovados:.2f}%")
```

## Questão 3 — Monitoramento de Consumo de Energia

(0,5 ponto)

Consumo diário (kWh):
`consumo = [120, 95, 200, 180, 75, 220, 160]`

Regras:
*   `< 100` → "Baixo"
*   `100–180` → "Moderado"
*   `> 180` → "Alto"

Tarefas:
*   Filtrar dias com consumo alto
*   Calcular:
    *   Total de consumo da semana
    *   Média semanal
*   Exibir alerta se houver mais de 2 dias com consumo alto

```python
consumo = [120, 95, 200, 180, 75, 220, 160]
classificacoes_consumo = []
dias_alto_consumo = 0

for consumo_diario in consumo:
    if consumo_diario < 100:
        classificacoes_consumo.append("Baixo")
    elif 100 <= consumo_diario <= 180:
        classificacoes_consumo.append("Moderado")
    else: # consumo_diario > 180
        classificacoes_consumo.append("Alto")
        dias_alto_consumo += 1

total_consumo_semana = sum(consumo)
media_consumo_semana = total_consumo_semana / len(consumo)

print(f"Consumo diário original: {consumo}")
print(f"Classificações de consumo: {classificacoes_consumo}")
print(f"Total de consumo da semana: {total_consumo_semana} kWh")
print(f"Média semanal de consumo: {media_consumo_semana:.2f} kWh")

if dias_alto_consumo > 2:
    print(f"ALERTA: Houve {dias_alto_consumo} dias com consumo alto, o que excede o limite de 2 dias.")
```

```python
precos = [50, 120, 30, 200, 80, 15]
precos_finais = []
total_com_desconto = 0
economia_total = 0

for preco_original in precos:
    desconto = 0
    if preco_original < 50:
        desconto = 0.05
    elif 50 <= preco_original <= 150:
        desconto = 0.10
    else: # preco_original > 150
        desconto = 0.15

    valor_desconto = preco_original * desconto
    preco_com_desconto = preco_original - valor_desconto

    precos_finais.append(f"{preco_com_desconto:.2f}")
    total_com_desconto += preco_com_desconto
    economia_total += valor_desconto

print(f"Preços originais: {precos}")
print(f"Preços finais com desconto: {precos_finais}")
print(f"Total da compra com desconto: {total_com_desconto:.2f}")
print(f"Economia total obtida: {economia_total:.2f}")
```

##  Questão 4 - Simulação de Carrinho de Compras



*   **Preços originais:** `[50, 120, 30, 200, 80, 15]`
*   **Regras de desconto aplicadas:**
    *   Itens com preço `< 50` receberam 5% de desconto.
    *   Itens com preço `50–150` receberam 10% de desconto.
    *   Itens com preço `> 150` receberam 15% de desconto.

*   **Preços finais com desconto:** `['45.00', '108.00', '28.50', '170.00', '72.00', '14.25']`
*   **Total da compra com desconto:** `437.75`
*   **Economia total obtida:** `57.25`




```python
precos = [50, 120, 30, 200, 80, 15]
precos_finais = []
total_com_desconto = 0
economia_total = 0

for preco_original in precos:
    desconto = 0
    if preco_original < 50:
        desconto = 0.05
    elif 50 <= preco_original <= 150:
        desconto = 0.10
    else: # preco_original > 150
        desconto = 0.15

    valor_desconto = preco_original * desconto
    preco_com_desconto = preco_original - valor_desconto

    precos_finais.append(f"{preco_com_desconto:.2f}")
    total_com_desconto += preco_com_desconto
    economia_total += valor_desconto

print(f"Preços originais: {precos}")
print(f"Preços finais com desconto: {precos_finais}")
print(f"Total da compra com desconto: {total_com_desconto:.2f}")
print(f"Economia total obtida: {economia_total:.2f}")
```
